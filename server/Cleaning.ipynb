## 2022-2023 Season Data Cleaning

```python
df2 = pd.read_csv('./data/matches-2022-2023.csv')
df2.head()
```

```python
df2 = df2[['date','clock','Home Team','Goals Home','Away Team','Away Goals']]
df2.head()
```

```python
df2['date']
```

```python
df2['date'] = pd.to_datetime(df2['date'], format='%d %B %Y', dayfirst=True)
```

```python
df2['date']
```

```python
# Sort the DataFrame by date
df2 = df2.sort_values(by='date').reset_index(drop=True)

# Assign game week numbers (assuming 38 game weeks)
df2['Game Week'] = (df2.index // (len(df2) // 38)) + 1
```

```python
df2['Game Week']
```

```python
df2.head()
```

```python
df2.sort_values(by="Home Team")
```

```python
def convert_to_24hr_format(clock):
    # Check if 'pm' or 'am' is in the clock
    if 'pm' in clock:
        time = clock.replace('pm', '').strip()
        if time.startswith('12'):
            time = time
        else:
            hours, minutes = time.split(':')
            time = f"{int(hours) + 12}:{minutes}"
    else:
        time = clock.replace('am', '').strip()
        if time.startswith('12'):
            time = f"00:{time.split(':')[1]}"
        else:
            time = time
    
    # Format time to 'HHMM'
    hours, minutes = time.split(':')
    return f"{hours}{minutes}"

df2['Time'] = df2['clock'].apply(convert_to_24hr_format)

df2['Date'] = pd.to_datetime(df2['date']).dt.strftime('%Y-%m-%d')

def get_match_result(row):
    if row['Goals Home'] > row['Away Goals']:
        return 'H'
    elif row['Goals Home'] < row['Away Goals']:
        return 'A'
    else:
        return 'D'

df2['Match Result'] = df2.apply(get_match_result, axis=1)

# Creating the final DataFrame
df_final = df2[['Time', 'Game Week', 'Date', 'Home Team', 'Away Team', 'Match Result']]

# Rename columns to match desired output format
df_final.columns = ['Time', 'Round', 'Date', 'Home Team', 'Away Team', 'Match Result']
```

```python
df_final['Time'].unique()
```

```python
def adjust_time_column(row):
    if row['Time'] == '0030p':
        return '1230'
    elif row['Time'] == '0000p':
        return '1200'
    else:
        return row['Time']
    
```

```python
df_final['Time'] = df_final.apply(adjust_time_column,axis=1)
```

```python
df_final['Date']
```

```python
df_final.info()
```

```python
df_final['Date'] = pd.to_datetime(df_final['Date'],format='%Y-%m-%d')
```

```python
df_final.info()
```

```python
df_final['Day'] = df_final['Date'].dt.day_name()
```

```python
df_final.drop(columns=['Date'],inplace=True)
```

```python
df_final.to_csv('./data/formatted/2022-2023.csv',index=False)
```
